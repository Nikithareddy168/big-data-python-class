```python
max_wins = data_frame[data_frame['attacker_outcome'] == "win"] ## Counts the total number of wins by each attacker
max_wins['attacker_king'].value_counts()
```

```python
## Plot for comparisions between different attackers about their winning battles
import matplotlib.pyplot as plt
ax = max_wins['attacker_king'].value_counts().plot(kind='bar', figsize=(10, 10), fontsize=12)
ax.set_title('Attack outcomes comparisions', fontsize=20)
ax.set_ylabel("Total Wins", fontsize=12)
plt.show()
```

```python
## sum of major deaths and major captures occured in each region

data = data_frame.groupby("region").sum()[["major_death", "major_capture"]]
print data
```
